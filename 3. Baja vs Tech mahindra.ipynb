# Rolling, Shifting and Moving Averages

```python
data = pd.read_csv('Bajaj_Auto.csv')  
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

data1 = pd.read_csv('Tech_Mahindra.csv')  
data1['Date'] = pd.to_datetime(data1['Date'])
data1.set_index('Date', inplace=True)

rolling_window = 8

# Calculate the rolling mean
data['Rolling Mean'] = data['Open'].rolling(window=rolling_window).mean()
data1['Rolling Mean'] = data1['Open'].rolling(window=rolling_window).mean()


# Plot the original closing prices, rolling mean, and rolling standard deviation
plt.figure(figsize=(12, 6))

plt.plot(data.index, data['Open'], label='Bajaj Auto Opening Price', color='deeppink', alpha=0.7)
plt.plot(data.index, data['Rolling Mean'], label='Bajaj Auto Rolling Mean (1 week)', color='forestgreen', linestyle='--')

plt.plot(data1.index, data1['Open'], label='Tech Mahindra Opening Price', color='indigo', alpha=0.7)
plt.plot(data1.index, data1['Rolling Mean'], label='Tech Mahindra Rolling Mean (1 week)', color='cyan', linestyle='--')

plt.title('Comparision of Stock Opening Prices with Rolling Mean of Bajaj Auto vs Tech Mahindra (Day Wise)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

# Expanding

```python
data['Expanding Mean (Opening)'] = data['Open'].expanding().mean()
data1['Expanding Mean (Opening)'] = data1['Open'].expanding().mean()


# Plot the original opening prices and expanding mean for opening prices
plt.figure(figsize=(12, 6))

plt.plot(data.index, data['Open'], label='Bajaj Auto Opening Price', color='indigo', alpha=0.7)
plt.plot(data.index, data['Expanding Mean (Opening)'], label='Bajaj Auto Expanding Mean (Opening)', color='red', linestyle='--')

plt.plot(data1.index, data1['Open'], label='Tech Mahindra Opening Price', color='darkgoldenrod', alpha=0.7)
plt.plot(data1.index, data1['Expanding Mean (Opening)'], label='Tech Mahindra Expanding Mean (Opening)', color='red', linestyle='--')


plt.title('Opening Prices with Expanding Mean of Bajaj Auto vs Tech Mahindra')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
data['Expanding Mean (Closing)'] = data['Close'].expanding().mean()
data1['Expanding Mean (Closing)'] = data1['Close'].expanding().mean()


# Plot the original opening prices and expanding mean for opening prices
plt.figure(figsize=(12, 6))

plt.plot(data.index, data['Close'], label='Bajaj Auto Opening Price', color='blueviolet', alpha=0.7)
plt.plot(data.index, data['Expanding Mean (Closing)'], label='Bajaj Auto Expanding Mean (Opening)', color='red', linestyle='--')

plt.plot(data1.index, data1['Close'], label='Tech Mahindra Opening Price', color='teal', alpha=0.7)
plt.plot(data1.index, data1['Expanding Mean (Closing)'], label='Tech Mahindra Expanding Mean (Opening)', color='red', linestyle='--')

plt.title('Opening Prices with Expanding Mean of Bajaj Auto vs Tech Mahindra')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
data = pd.read_csv('Bajaj_Auto.csv') 

# Convert the 'Date' column to a datetime object
data['Date'] = pd.to_datetime(data['Date'])

# Set the 'Date' column as the DataFrame index
data.set_index('Date', inplace=True)

# Plot the 'High' and 'Low' prices over time
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['High'], label='High Price', color='green', alpha=0.7)
plt.plot(data.index, data['Low'], label='Low Price', color='red', alpha=0.7)
plt.title('High and Low Prices Over Time for Bajaj Auto')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Resample the data on a weekly basis and calculate the weekly high and low prices
weekly_data = data.resample('W').agg({'High': 'max', 'Low': 'min'})

# Plot the weekly high and low prices
plt.figure(figsize=(12, 6))
plt.plot(weekly_data.index, weekly_data['High'], label='Weekly High Price', color='green', alpha=0.7)
plt.plot(weekly_data.index, weekly_data['Low'], label='Weekly Low Price', color='red', alpha=0.7)
plt.title('Weekly High and Low Prices for Bajaj Auto')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
