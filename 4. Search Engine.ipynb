### Preprocessing

```python
stop_words = stopwords.words('english')
stop_words
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
vectorizer
```

```python
document_vectors = vectorizer.fit_transform(documents)
document_vectors
```

### User Search Query 

```python
query = "Bajaj Auto stock performance"

# We can edit the search query as per our need
```

### Preprocess and vectorize query

```python
query_vector = vectorizer.transform([query])
```

### Calculate cosine similarity between query vector and document vectors

```python
similarities = cosine_similarity(query_vector, document_vectors).flatten()
similarities
```

### Rank documents based on similarity

```python
ranked_documents = sorted(zip(similarities, documents), reverse=True)
ranked_documents
```

### Print top 3 ranked documents

```python
for rank, (similarity, document) in enumerate(ranked_documents[:3], start=1):
    print(f"Rank {rank}: Similarity {similarity:.2f}, Document: {document}")
```
