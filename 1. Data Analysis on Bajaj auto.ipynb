```python
df = pd.read_csv("Bajaj_Auto.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
df.isnull().sum()
```

```python
series = df['Open']

# 1. Visual the output variable 
plt.plot(series)
plt.title('Open Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
# 2. Augmented Dickey-Fuller (ADF) Test
adf_result = adfuller(series)
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
print('Critical Values:', adf_result[4])
```

```python
# data is non stationary lets try to do make it stattionary 
```

```python
# 3. Differencing
series_diff = series.diff().dropna()

# 4. Repeat ADF Test
adf_result_diff = adfuller(series_diff)
print('\nADF Statistic (after differencing):', adf_result_diff[0])
print('p-value (after differencing):', adf_result_diff[1])
print('Critical Values (after differencing):', adf_result_diff[4])
```

```python
df = pd.read_csv("Bajaj_Auto.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
series = df['Open']

# seasonal decomposition
decomposition = seasonal_decompose(series, model='additive', period=1)  # Adjust period if seasonality is known

# Plot decomposition
plt.figure(figsize=(12, 8))

plt.subplot(411)
plt.plot(series, label='Original')
plt.legend()

plt.subplot(412)
plt.plot(decomposition.trend, label='Trend')
plt.legend()

plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonal')
plt.legend()

plt.subplot(414)
plt.plot(decomposition.resid, label='Residual')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
series = df['Open']

order = (1, 1, 1)          # Non-seasonal parameters (p, d, q)
seasonal_order = (1, 1, 1, 12)  # Seasonal parameters (P, D, Q, m)

model = SARIMAX(series, order=order, seasonal_order=seasonal_order)
results = model.fit()

forecast = results.forecast(steps=3)

print("Predicted Open Prices for Next 3 Days:")
print(forecast)
```
